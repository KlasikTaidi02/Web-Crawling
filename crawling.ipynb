```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
sitkul = urlopen('https://kuliah.uajy.ac.id/')
sitkul_html = soup(sitkul.read(), 'html.parser')
print(sitkul_html)
```

```python
daftar_link=[]

for link in sitkul_html.find_all('a'):
    dic = {}
    if'href' in link.attrs:
        dic = link.get('href')
        daftar_link.append(dic)
        

daftar_link_df = pd.DataFrame(daftar_link)
```
