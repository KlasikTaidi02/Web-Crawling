```python
import requests
import pandas as pd
import html5lib

from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population'

response = requests.get(url)    
soup = BeautifulSoup(response.text, "html.parser")

```

```python
print(soup)
```

```python
rows = soup.find('table').find('tbody').find_all('tr')

print(rows)

print(len(rows))
```

```python
countries_list = []



for row in rows :
    dic = {}
    td_elements = row.find_all('td')
   
    # Check if there are elements in the list before accessing them
    if len(td_elements) > 0:
        dic['Country'] = td_elements[0].text
        dic['Population'] = td_elements[1].text
        dic["'% 'of' World"] = td_elements[2].text
        dic['Date'] = td_elements[3].text
        dic['Source'] = td_elements[4].find('a').get('href')
        countries_list.append(dic)

print(countries_list)




```

```python
Countries = pd.DataFrame(countries_list)

Countries.to_excel('countries_data.xlsx',index = False)
Countries.to_csv('countries_data.csv', index = False)
```
